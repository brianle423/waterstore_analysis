```python
import pandas as pd
import numpy as np
import warnings
from glob import glob

warnings.filterwarnings('ignore')
pd.options.plotting.backend = "plotly"
pd.set_option("display.max_columns", None)
```

# Lucky LTD

```python
lucky = pd.read_excel("../Water Daily Sales/Raw Data/Year 2021/TwoStore_DailySale Jun-21.xlsx", sheet_name='LTD_Lucky')
```

```python
lucky = lucky.drop(columns='Unnamed: 0')
lucky = lucky.set_index('P$L')
```

```python
lucky.head()
```

```python
lucky = lucky.transpose()
```

```python
lucky_expenses = lucky[['Merchandise ',
 'Maintenance',
 'Salaries',
 'Insurance WC',
 'Payroll Tax',
 'Rent',
 'Water',
 'PG&E',
 'Phone',
 'Insurance',
 'Bank Charge',
 'Office, Cleaning Supplies',
 'Repair & Maintenance',
 'Marketing Expenses',
 'License & Permit',
 'Property Tax ',
 '    Total Expenses']]
lucky_expenses = lucky_expenses.fillna(0)
expense = ['Merchandise ',
 'Maintenance',
 'Salaries',
 'Insurance WC',
 'Payroll Tax',
 'Rent',
 'Water',
 'PG&E',
 'Phone',
 'Insurance',
 'Bank Charge',
 'Office, Cleaning Supplies',
 'Repair & Maintenance',
 'Marketing Expenses',
 'License & Permit',
 'Property Tax ']
lucky_expenses['Total Expenses'] = lucky_expenses.loc[: , expense].sum(axis=1)
lucky_expenses = lucky_expenses[:-6]
```

```python
lucky_expenses = lucky_expenses.reset_index()
lucky_expenses = lucky_expenses.rename({'index':'Date', 'Merchandise ':'Merchandise','Property Tax ':'Property Tax'}, axis=1)
lucky_expenses['Date'] = pd.to_datetime(lucky_expenses['Date'])
lucky_expenses = lucky_expenses.set_index('Date')
```

```python
lucky_expenses = lucky_expenses.drop('    Total Expenses', axis=1)
```

```python
lucky_expenses = lucky_expenses.replace(' -   ', 0)
```

```python
fm = pd.read_excel("../Water Daily Sales/Raw Data/Year 2021/TwoStore_DailySale Jun-21.xlsx", sheet_name='LTD_FM')
```

```python
fm = fm.drop(columns='Unnamed: 0')
fm = fm.set_index('P$L')
```

```python
fm = fm.transpose()
```

```python
fm.columns.to_list()
```

```python
fm_expenses = fm[['Merchandise ',
 'Salt & Filters',
 'Salaries',
 'Insurance WC',
 'Payroll tax',
 'Rent',
 'Water',
 'PG&E',
 'Phone',
 'Insurance',
 'Bank Charge',
 'Office, Cleaning Supplies',
 'Repair & Maintenance',
 'Marketing Expenses',
 'License & Permit',
 'Property Tax ',
 '    Total Expenses']]
fm_expenses = fm_expenses.fillna(0)
```

```python
cols = ['Merchandise ',
 'Salt & Filters',
 'Salaries',
 'Insurance WC',
 'Payroll tax',
 'Rent',
 'Water',
 'PG&E',
 'Phone',
 'Insurance',
 'Bank Charge',
 'Office, Cleaning Supplies',
 'Repair & Maintenance',
 'Marketing Expenses',
 'License & Permit',
 'Property Tax ']
```

```python
fm_expenses['Total Expenses'] = fm_expenses.loc[: , cols].sum(axis=1)
fm_expenses = fm_expenses[:-6]
```

```python
fm_expenses = fm_expenses.reset_index()
fm_expenses = fm_expenses.rename({'index':'Date', 'Merchandise ':'Merchandise','Property Tax ':'Property Tax'}, axis=1)
fm_expenses['Date'] = pd.to_datetime(fm_expenses['Date'])
fm_expenses = fm_expenses.set_index('Date')
```

```python
fm_expenses.dtypes
```

```python
fm_expenses = fm_expenses.replace([' -   ', ' '], 0)
```

```python
fm_expenses = fm_expenses.drop('    Total Expenses', axis=1)
```
